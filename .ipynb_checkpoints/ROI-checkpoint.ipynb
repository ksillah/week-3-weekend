```python
from IPython.display import clear_output
class Prop:
    def __init__(self,value,income,expenses,taxes,insurance,utilities,
                 hoa,landscaping,vacency,repairs,capx,manager,mortgage,flow,closing,rehab,invest,roi):
        self.value = value
        self.income = income
        self.expenses = expenses
        self.taxes = taxes
        self.insurance = insurance
        self.utilities = utilities
        self.hoa = hoa
        self.landscaping = landscaping
        self.vacency = vacency
        self.repairs = repairs
        self.capx = capx
        self.manager = manager
        self.mortgage = mortgage
        self.flow= flow
        self.closing = closing
        self.rehab = rehab
        self.invest = invest
        self.roi = roi
    def findIncome(self):
        while True:
            try:
                rent =int(input('What will be your monthly rental income? Please input a number.'))
                more_income = input('Would you like to include more monthly income? Please respond with yes or no.')
                clear_output()
                if more_income.lower() == 'yes':
                    more_input = int(input('Please input your additional monthly income. Please input a number.'))
                    if type(more_input) == int:
                        self.income=( rent + more_input)
                        print(self.income)
                        return False
                    else:
                        print('Please try again')
                elif more_income.lower() == 'no':
                    self.income=rent
                    return False
                else:
                    print('Please try again')
            except:
                print('Please try again')
                
    def findExpenses(self):
        while True:
            try:
                exp =input('Do you want to input your monthly expenses or assume the average expenses in Illinois? /n Respond with "M" to include your own input.')
                if exp.lower() == 'm':
                    self.taxes=int(input(("How much will your monthly property tax cost?")))
                    self.insurance =int(input(("How much will your monthly insurance cost?")))
                    self.utilities = int(input(("How much will your monthly utilities cost?")))
                    self.hoa = int(input(("How much will your monthly HOA fees cost?")))
                    self.landscaping = int(input(("How much will your monthly lawn care cost?")))
                    self.vacency = int(input(("How much will you allocate monthly for vacency?")))
                    self.repairs = int(input(("How much will you allocate monthly for repairs?")))
                    self.capx = int(input(("How much will you allocate monthly for capital expenditures?")))
                    self.manager = int(input(("How much will you allocate monthly for a property manager?")))
                    self.mortgage = int(input(("How much will you allocate monthly for mortgage?")))
                    self.expenses = (self.taxes+self.insurance+self.utilities+self.hoa+self.landscaping+self.vacency+self.repairs+self.capx+self.manager+self.mortgage)
                    print(self.expenses)
                    return False
                else:
                    return False
            except:
                print("Please try again.")
    def findInvestment(self):
         while True:
            clear_output()
            try:
                self.down =int(input('How much was your downpayment? Please input a number.'))
                self.closing= int(input('How much were your closing costs? Please input a number.'))
                self.rehab = int(input('What was your rehab budget?'))
                self.rehab+= int(input('Input the amount of additional expenses'))
                self.invest = self.down + self.closing + self.rehab
                self.roi = round((self.flow/self.invest)*100,2)
                return False
            except:
                print('Please try again')

            
def run():
    try:
        value=int(input("Welcome to Kutah's ROI calculator! What is the value of your property? Please insert a number."))
        taxes=round(.0227*value/12,2)
        closing=.02*value
        down=.2*value
        invest = down+closing
        your_property=Prop(value,0,1500,taxes,100,100,0,0,100,100,100,0,1000,down,closing,0,invest,0)
        your_property.findIncome()
        your_property.findExpenses()
        your_property.flow=(your_property.income-your_property.expenses)*12
        your_property.findInvestment()
        clear_output()
    
        print(f'Your return on investment is {your_property.roi} percent.')
        print(f"Input Receipt: All values are monthly unless otherwise noted \n\nProperty value: {your_property.value}\nIncome Tax: {your_property.income}\nTotal Expenses: {your_property.expenses}")
        print(f"Property Taxes: {your_property.taxes}\nProperty Insurance: {your_property.insurance}\nUtilities: {your_property.utilities}")
        print(f"HOA Fees: {your_property.hoa} \nLandscaping Fees: {your_property.landscaping} \nVacency Fees: {your_property.vacency}")
        print(f"Repair Costs: {your_property.repairs}\nCapital Expenditures: {your_property.capx} \nProperty Management: {your_property.manager}")
        print(f"Annual Cash Flow: {your_property.flow}\nClosing Costs: {your_property.closing}\nRehab costs: {your_property.rehab}")
        print(f"Total Investment: {your_property.invest}\nReturn on Investment: {your_property.roi} percent")
    except:
        print("Please try again.")
    
run()
    
        
```
